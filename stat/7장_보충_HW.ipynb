```python
import numpy as np

# 설정
mu = 100  # 평균 수명
sigma = 10  # 표준편차
n = 30  # 전구 개수
days = 100  # 측정 일수

# 100일 동안 매일 30개의 전구 평균 수명 시뮬레이션
daily_averages = np.array([np.mean(np.random.normal(mu, sigma, n)) for _ in range(days)])

# 97~103 범위를 벗어난 횟수 계산
print(" daily averages : ", daily_averages[0])
out_of_range_count = np.sum((daily_averages < 97) | (daily_averages > 103))
daily_averages_under97 = daily_averages[daily_averages < 97]
daily_averages_over103 = daily_averages[daily_averages > 103]
print(len(daily_averages_under97))
print(len(daily_averages_over103))
print("out of range count : ", out_of_range_count)

# 100개의 평균값의 분포 확인을 위한 히스토그램 데이터 생성
histogram_data = np.histogram(daily_averages, bins=10, range=(np.min(daily_averages), np.max(daily_averages)))
print(np.min(daily_averages), np.max(daily_averages))
print("histogram_data[0] sum: ", sum(histogram_data[0]))
print("histogram_data : ", len(histogram_data[1]))
import matplotlib.pyplot as plt

# 히스토그램 그래프 그리기
plt.hist(daily_averages, bins=10, edgecolor='black')
plt.title('100 Days of Daily Average Lifespans of Bulbs')
plt.xlabel('Average Lifespan')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np

# 설정
lamda = 100  # 평균 수명
n = 30  # 전구 개수
days = 100  # 측정 일수

# 100일 동안 매일 30개의 전구 평균 수명 시뮬레이션
daily_averages = np.array([np.mean(np.random.exponential(lamda, n)) for _ in range(days)])

# 97~103 범위를 벗어난 횟수 계산
#print(" daily averages : ", daily_averages)
out_of_range_count = np.sum((daily_averages < 97) | (daily_averages > 103))
daily_averages_under97 = daily_averages[daily_averages < 97]
daily_averages_over103 = daily_averages[daily_averages > 103]
print(len(daily_averages_under97))
print(len(daily_averages_over103))
print("out of range count : ", out_of_range_count)

# 100개의 평균값의 분포 확인을 위한 히스토그램 데이터 생성
histogram_data = np.histogram(daily_averages, bins=10, range=(np.min(daily_averages), np.max(daily_averages)))
print(np.min(daily_averages), np.max(daily_averages))
print("histogram_data[0] sum: ", sum(histogram_data[0]))
print("histogram_data : ", len(histogram_data[1]))
import matplotlib.pyplot as plt

# 히스토그램 그래프 그리기
plt.hist(daily_averages, bins=10, edgecolor='black')
plt.title('100 Days of Daily Average Lifespans of Bulbs')
plt.xlabel('Average Lifespan')
plt.ylabel('Frequency')
plt.show()
```
